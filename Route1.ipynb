```

#Exercise 1
import pandas as pd
df=pd.read_csv('mtb_uniprot_subset.csv')
print(df.head(5))

#Exercise 2
print(df.info())
print(df.describe(include='all'))

#Exercise 3
print(df.query("Length > 30 & Reviewed.str.startswith('r').values"))
#df['Reviewed']=df['Reviewed'].astype('category')
#filteredcategory_df=df[df['Reviewed']==['reviewed']]
#filtered_df=df.query('Length > 30')
#print(filtered_df)

#Exercise 4
df.rename(columns={'Length':'number of AA', 'Entry':'Uniprot ID'}, inplace=True)
print(df.info())

#Exercise 5
print(len(df))
df.drop_duplicates(inplace=True)
df.drop_duplicates(inplace=True, subset=['Reviewed'])
print(df)
print(len(df))

#Exercise 6
missingcount=df.isna().sum()
df2=pd.read_csv('mtb_uniprot_subset.csv')
df2.fillna(0)
df2.fillna(df.median(numeric_only=True))
df2.fillna(df.mean(numeric_only=True))
df2.fillna(df.mode(numeric_only=True))

df.ffill(inplace=True)
df.bfill(inplace=True)

#Exercise 7
#finding name of biggest protein in dataset
biggestsizeindex=df2['Length'].idxmax()
biggestname=df2.loc[biggestsizeindex]
print(biggestname)

```
