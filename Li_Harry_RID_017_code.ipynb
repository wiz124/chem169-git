```python
#Exercise 0
import pandas as pd

df=pd.read_excel('mtb_with_localization.xlsx')
display(df)
```

```python
#Exercise 1

location=[]
for entry in df['Subcellular location [CC]']:
  if 'membrane' in str(entry).lower():
    location.append('membrane')
  elif 'secreted' in str(entry).lower():
    location.append('secreted')
  elif 'cytoplasm' in str(entry).lower():
    location.append(  'cytoplasm')
  else:
    location.append(None)

df['location']=location
df.dropna(subset=['location'], inplace=True)
display(df)

print(len(df[df['location'] == 'membrane']))
print(len(df[df['location'] == 'cytoplasm']))
print(len(df[df['location'] == 'secreted']))

```

```python
#Exercise 2
import numpy as np

length=[]
for sequence in df['Sequence']:
  length.append(len(sequence))

df['length']=length

print(df['length'].max())
print(df['length'].min())
print(df['length'].mean())

x = df[['length']].to_numpy()
y = df[['location']].to_numpy()

```

```python
#Exercise 3
def count_amino_acids(sequence):
    amino_acids = list('ACDEFGHIKLMNPQRSTVWY')
    sequence=str(sequence)
    return {aa: sequence.count(aa) for aa in amino_acids}

amino_acids = list('ACDEFGHIKLMNPQRSTVWY')

aa_counts = pd.DataFrame(
    df['Sequence'].apply(count_amino_acids).tolist(),
    columns=amino_acids,
    index=df.index
)
df = pd.concat([df, aa_counts], axis=1)



x = df[['length'] + amino_acids].to_numpy()
print(x.shape)
```

```python
#Exercise 4
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.3,      # 30% for testing
    random_state=42,    # for reproducibility
    stratify=y          # maintain class distribution
)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Exercise 5
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear',max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Exercise 6
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100}%")

```
