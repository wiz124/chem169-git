```python
#exercise 0
import pandas
df=pandas.read_csv('database.txt', skiprows=2, sep=';')
display(df)
print(df[' SMILES'])
```

```python
#Exercise 1
from rdkit import Chem
from rdkit.Chem.rdchem import Mol
def smilestomol(smile):
  try:
    mol=Chem.MolFromSmiles(smile)
    return mol
  except:
    return None
df['mol']=df[' SMILES'].apply(smilestomol)
display(df)
```

```python
#Exercise 2
from rdkit import Chem
from rdkit.Chem import Descriptors
def size(mol):
  if mol==None:
    return 0.0
  elif Chem.Descriptors.MolWt(mol)>0:
    mmass=Chem.Descriptors.MolWt(mol)
    return mmass
  else:
    return 0.0

df['molecular mass']=df['mol'].apply(size)
display(df)
```
